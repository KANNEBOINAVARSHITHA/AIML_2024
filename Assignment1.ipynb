Part-02:Lab Task is to visualizw the datasets or data using Matplot function Graphs

Generate Scatterplot

Let's suppose we collected age (in years),height(in cm), and weight(in kilograms) infomate visiting a clinic.

```python
people=['Ann','Brandom','Chen','Emily','Farook','Gagan','Hamish']
age=[15,20,17,45,50,22,30]
weight=[40,45,50,55,60,65,35]
height=[4.0,4.5,5.0,5.5,6.0,6.5,3.5]
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("Relationship between weight and Height of patients")
plt.ylabel("Height in meters")
plt.xlabel("Weight of patients in Kgs")
plt.show()
plt.figure(figsize=(12,4))
plt.title("People's weight in Kgs",fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight(in Kgs)",fontsize=15)
plt.show()
```

Generate Histogram


```python
import numpy as np
plt.figure(figsize=(7,5))
#Main plot function 'hist'
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in Kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
#Implement trignometric and log functions of a given set
import numpy as np
patientage=np.array([65,78,45,67,56,77])
sin1=np.sin(patientage)
cosine=np.cos(patientage)
Tangent=np.tan(patientage)
Log1=np.log10(patientage)
LOG=np.log2(patientage)
EXP=np.exp(patientage)
print("The sine of the patientage is",sin1)
print("The cosine of the patientage is",cosine)
print("The tangent of the patientage is",Tangent)
print("The logarithm base 2 of the patientage is",Log1)
print("The antilogarithm result of the patientage is",LOG)
print("The expotentional form of the patientage is",EXP)
```
